```python
import pandas as pd
from pybaseball import standings, season_batting_bref

season_standings = []

for season in range(2021, 2026):
  season_data = standings(season)
  season_data_combined = pd.concat(season_data, ignore_index=True)
  season_data_combined['season'] = season
  season_standings.append(season_data_combined)

team_batting_bref = season_batting_bref(2021, 2025)
```

```python
print(season_standings[0])
print('\nChecking output of team batting')
print(team_batting_bref.shape)
print()
print(team_batting_bref.head(n=1))
```

```python
filtered_team_batting = team_batting_bref[['season', 'team_abbrev', 'team_name', 'runs_per_game', 'onbase_plus_slugging', 'onbase_plus_slugging_plus']]
filtered_team_batting
```

```python
mlb_standings_2021_2025 = pd.concat(season_standings, ignore_index=True)
mlb_standings_2021_2025['GB'] = mlb_standings_2021_2025['GB'].replace('--', 0)
mlb_standings_2021_2025.to_csv('mlb_standings_2021_2025.csv', index=False)
filtered_team_batting.to_csv('team_batting_2021_2025.csv', index=False)
```
