```python
import pandas as pd
import numpy as np
from pybaseball import bwar_bat, bwar_pitch

bref_batting = bwar_bat()
bref_pitching = bwar_pitch()
```

```python
print(bref_batting.columns)
print(bref_pitching.columns)
```

```python
filtered_bwar_batting = bref_batting[(bref_batting['year_ID'] >= 2021) & (bref_batting['pitcher'] == 'N') & (bref_batting['PA'] >= 100)]
filtered_bwar_batting = filtered_bwar_batting[['name_common', 'mlb_ID', 'year_ID', 'team_ID', 'lg_ID', 'WAR']]
filtered_bwar_batting
```

```python
filtered_bref_pitching = bref_pitching.copy(deep=True)
filtered_bref_pitching = filtered_bref_pitching[filtered_bref_pitching['year_ID'] >= 2021]
filtered_bref_pitching[['name_common', 'IPouts']].head(5)
```

```python
filtered_bref_pitching['IPouts'] = filtered_bref_pitching['IPouts'] // 3 + ((filtered_bref_pitching['IPouts'] % 3) * .1)
filtered_bref_pitching[['name_common', 'IPouts']].head(5)
```

```python
filtered_bwar_pitching = filtered_bref_pitching[filtered_bref_pitching['IPouts'] >= 20]
filtered_bwar_pitching = filtered_bwar_pitching[['name_common', 'mlb_ID', 'year_ID', 'team_ID', 'lg_ID', 'WAR']]
filtered_bwar_pitching
```

```python
filtered_bwar_pitching.to_csv('bwar_pitching_data.csv', index=False)
filtered_bwar_batting.to_csv('bwar_batting_data.csv', index=False)
```
