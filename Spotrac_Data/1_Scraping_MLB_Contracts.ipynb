```python
import pandas as pd
import numpy as np

from mlb_contracts_scraper import season_contracts
```

```python
contracts_2021 = season_contracts(2021)
contracts_2022 = season_contracts(2022)
contracts_2023 = season_contracts(2023)
contracts_2024 = season_contracts(2024)
contracts_2025 = season_contracts(2025)
```

```python
contracts_2021 = pd.read_csv('2021_mlb_contracts.csv')
contracts_2022 = pd.read_csv('2022_mlb_contracts.csv')
contracts_2023 = pd.read_csv('2023_mlb_contracts.csv')
contracts_2024 = pd.read_csv('2024_mlb_contracts.csv')
contracts_2025 = pd.read_csv('2025_mlb_contracts.csv')
```

```python
all_contracts = pd.concat([contracts_2021, contracts_2022, contracts_2023, contracts_2024, contracts_2025], ignore_index=True)
all_contracts
```

```python
all_contracts = all_contracts.rename(columns={
  'Player': 'player',
  'Pos': 'position',
  'Age': 'age',
  'Type': 'contract_type',
  'Luxury Tax': 'luxury_tax',
  'Luxury Tax %': 'luxury_tax_pct',
  'Cash Salary': 'salary',
  'Free Agent Year': 'free_agency_yr'
})

all_contracts = all_contracts.replace(np.nan, None)
all_contracts = all_contracts.replace('-', None)
all_contracts = all_contracts.replace('Free AgentForeign Professional', 'Free Agent')
all_contracts = all_contracts.replace('ExtensionArbitration', 'Extension')
all_contracts = all_contracts.replace('ExtensionPre-Arbitration', 'Extension')

all_contracts['luxury_tax'] = all_contracts['luxury_tax'].str.replace(',', '')
all_contracts['luxury_tax'] = all_contracts['luxury_tax'].str.replace('$', '')
all_contracts['luxury_tax'] = pd.to_numeric(all_contracts['luxury_tax'])
all_contracts['luxury_tax'] = all_contracts['luxury_tax'].replace(np.nan, 0)

all_contracts['salary'] = all_contracts['salary'].str.replace(',', '')
all_contracts['salary'] = all_contracts['salary'].str.replace('$', '')
all_contracts['salary'] = pd.to_numeric(all_contracts['salary'])
all_contracts['salary'] = all_contracts['salary'].replace(np.nan, 0)

all_contracts['luxury_tax_pct'] = all_contracts['luxury_tax_pct'].str.replace('%', '')
all_contracts['luxury_tax_pct'] = pd.to_numeric(all_contracts['luxury_tax_pct']) / 100
all_contracts['luxury_tax_pct'] = all_contracts['luxury_tax_pct'].replace(np.nan, 0)

all_contracts = all_contracts.replace(np.nan, None)

all_contracts
```

```python
all_contracts.to_csv('mlb_contracts_2021_2025.csv', index=False)
```
